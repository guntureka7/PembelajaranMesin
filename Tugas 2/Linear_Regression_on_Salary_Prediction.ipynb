```python
# Langkah 1: Impor library
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Langkah 2: Baca dataset
data = pd.read_csv("/content/drive/MyDrive/dataset/Salary_Data_Simple_SVR.csv")

# Langkah 3: Pisahkan dataset menjadi fitur (X) dan target (y)
X = data.iloc[:, :-1].values
y = data.iloc[:, 1].values

# Langkah 4: Bagi dataset menjadi data pelatihan dan data pengujian
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Langkah 5: Buat model regresi linear
regressor = LinearRegression()

# Langkah 6: Latih model
regressor.fit(X_train, y_train)

# Langkah 7: Evaluasi model
y_pred = regressor.predict(X_test)

# Langkah 8: Visualisasikan hasil prediksi
plt.scatter(X_test, y_test, color='red')
plt.plot(X_test, y_pred, color='blue')
plt.title('Gaji vs Pengalaman (Data Pengujian)')
plt.xlabel('Pengalaman (Tahun)')
plt.ylabel('Gaji ($)')
plt.show()
```

```python
# Visualisasi data pelatihan
plt.scatter(X_train, y_train, color='red', label='Data Pelatihan')

# Visualisasi data pengujian
plt.scatter(X_test, y_test, color='blue', label='Data Pengujian')

# Visualisasi garis regresi
plt.plot(X_train, regressor.predict(X_train), color='green', label='Garis Regresi')

plt.title('Gaji vs Pengalaman')
plt.xlabel('Pengalaman (Tahun)')
plt.ylabel('Gaji ($)')
plt.legend()
plt.show()
```
